```python
rgb_maze= plt.imread(maze)

plt.figure(figsize=(7, 7))
plt.imshow(rgb_maze)

x0,y0= 190, 5
x1,y1= 210, 410

plt.plot(x0,y0, 'rx', markersize = 14)
plt.plot(x1,y1, 'bx', markersize = 14)

# Add labels
plt.text(x0 + 5, y0, 'Start', color='red', fontsize=12)
plt.text(x1 + 5, y1, 'Goal', color='blue', fontsize=12)

# Add title and labels
plt.title('Maze with Entry and Goal Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show legend
plt.legend()
```
